```python
# Class Student 선언
class Student:
    
    # 객체생성 (field)
    def __init__(self, name=None, kor = None, eng = None, math = None):
        self.__name = name
        self.__kor = kor
        self.__eng = eng
        self.__math = math
        self.__total = self.__kor + self.__eng + self.__math
        self.__avg = '%.1f' % (self.__total/3)

    # 학생정보 입력
#     def setInfo(self):
#         self.name = input('학생 이름을 입력하세요 : ')
#         self.kor =  int(input('국어 성적을 입력하세요 : '))
#         self.eng =  int(input('영어 성적을 입력하세요 : '))
#         self.math = int(input('수학 성적을 입력하세요 : '))
#         print('성적이 입력이 완료 되었습니다.\n')
        
#         self.total = self.kor + self.eng + self.math
#         self.avg = '%.1f' % (self.total/3)
    
    def setName(self, name):
        self.__name = name
    def setKor(self, kor):
        self.__kor = kor
    def setEng(self, eng):
        self.__eng = eng
    def setMath(self, math):
        self.__math = math
    def setAutoCalc(self):
        self.__total = self.kor + self.eng + self.math
        self.__avg = '%.1f' % (self.total/3)
        
    # 학생정보 출력
#     def getInfo(self):
#         print(self.name, '\t', self.kor, '\t', self.eng, '\t', self.math, '\t', self.total, '\t', self.avg)
    
    # 학생정보 출력2
    def getName(self):
        return self.__name
    def getKor(self):
        return self.__kor
    def getEng(self):
        return self.__eng
    def getMath(self):
        return self.__math
```

```python
# 학생 정보를 table에 info Class를 이용하여 입력
table = list()     # 학생 정보가 입력된 Class가 모여있음
namelist = list()  # 각 Class에 입력된 학생 이름만 모아놓음

   
def printResult(i):
    name = table[i].getName()
    kor = table[i].getKor() 
    eng = table[i].getEng()
    math = table[i].getMath()
    total = int(kor) + int(eng) + int(math)
    avg = '%.1f' % (total/3)

    print(name, '\t', kor, '\t', eng, '\t', math, '\t', total, '\t', avg)
    
def inputInfo():
    print('\n>[1.입력] 학생정보 입력을 시작합니다')
    
    name = input('학생 이름을 입력하세요 : ')
    
    while (name in namelist) == True:
        print('이미 입력된 학생이름입니다. 이름을 변경하여 입력하세요')
        name = input('학생 이름을 입력하세요 : ')
    
    kor =  int(input('국어 성적을 입력하세요 : '))
    eng =  int(input('영어 성적을 입력하세요 : '))
    math = int(input('수학 성적을 입력하세요 : '))
    
    info = Student(name, kor, eng, math)    # 새로운 Class호출
    tmpId = len(table)  # table list의 index 위치

    table.append(info)
    #table[tmpId].setInfo()

    # 이름list 업데이트
    namelist.append(table[tmpId].getName())

    print('성적이 입력이 완료 되었습니다.\n')
    print('\n[1.입력] 입력 결과를 출력합니다')
    print('이름\t국어\t영어\t수학\t합계\t평균')
    print('-'*50)
    printResult(tmpId)

    print('\n[1.입력] 현재 입력된 학생 리스트 : ', namelist)

    
def outputInfo():
    if len(table) == 0:
        print('\n>[2.출력] 입력된 학생정보가 없습니다')
    else:
        print('\n>[2.출력] 학생정보를 출력합니다')
        print('이름\t국어\t영어\t수학\t합계\t평균')
        print('-'*50)

        for i in range(len(table)):
            printResult(i)

        print('>[2.출력] fin')

def searchInfo():
    nameSearch = input('\n>[3.검색] 성적을 검색할 학생의 이름을 입력하세요: ')

    if (nameSearch in namelist) == False:
        print()
        print("'",nameSearch,"' 라는 학생은 입력된 성적이 없습니다")

    else:
        tmpId = namelist.index(nameSearch) # table list의 index 위치

        print('\n>[3.검색] 검색된 학생정보를 출력합니다')
        print('이름\t국어\t영어\t수학\t합계\t평균')
        print('-'*50)
        printResult(tmpId)

def modiInfo():
    nameSearch = input('\n>[4.수정] 성적을 수정할 학생의 이름을 입력하세요: ')

    if (nameSearch in namelist) == False:
        print()
        print("\n>[4.수정] '",nameSearch,"' 라는 학생은 입력된 성적이 없습니다")
    else:
        tmpId = namelist.index(nameSearch) # table list의 index 위치
        print('1.국어 2.영어 3.수학')
        sub = input('\n>[4.수정] 성적을 수정할 과목 번호를 입력하세요')

        if sub == '1':
            score = int(input('국어 성적을 입력하세요 : '))
            table[tmpId].setKor(score)
            table[tmpId].setAutoCalc
                      
        elif sub == '2':
            score = int(input('영어 성적을 입력하세요 : '))
            table[tmpId].setEng(score)
            table[tmpId].setAutoCalc
                        
        elif sub == '3':
            score = int(input('수학 성적을 입력하세요 : '))
            table[tmpId].setMath(score)
            table[tmpId].setAutoCalc
                           
        else:
            print('\n>[4.수정] 과목 번호를 잘못입력하였습니다')
        
        print('\n수정후 데이터를 확인하세요')
        print('이름\t국어\t영어\t수학\t합계\t평균')
        print('-'*50)
        printResult(tmpId)

def delInfo():
    nameSearch = input('\n>[5.삭제] 성적을 삭제할 학생의 이름을 입력하세요: ')

    if (nameSearch in namelist) == False:
        print()
        print("\n>[5.삭제] '",nameSearch,"' 라는 학생은 입력된 성적이 없습니다")

    else:
        tmpId = namelist.index(nameSearch) # table list의 index 위치

        print('\n>[5.삭제] 다음 학생정보를 삭제하겠습니다')
        print('이름\t국어\t영어\t수학\t합계\t평균')
        print('-'*50)
        printResult(tmpId)

        table.pop(tmpId)
        namelist.pop(tmpId)
        print('\n>[5.삭제] 삭제가 완료되었습니다')
        print('\n>[5.삭제] 삭제 후 입력된 학생 리스트 : ', namelist)  
```

```python
##### Class 호출
# info = Student()

# 반복실행
while True:
    print()
    print('*'*50)
    print('1.입력 2.출력 3.검색 4.수정 5.삭제 6.종료')
    menu = input('실행을 원하는 메뉴 번호를 입력하세요 : ')
    print('*'*50)

    # 1.입력
    if menu == '1':
        inputInfo()
    
    # 2.출력
    elif menu == '2':
        outputInfo()
    
    # 3.검색
    elif menu == '3':
        searchInfo()
    
    # 4.수정
    elif menu == '4':
        modiInfo()
    
    # 5.삭제
    elif menu == '5':
        delInfo()
        
    # 6.종료
    elif menu == '6':
        print('')
        print('\n>[6.종료] 프로그램을 종료합니다')
        break
    
    # menu 입력오류
    else:
        print('')
        print('\n메뉴 번호를 잘못 입력하였습니다')
```
