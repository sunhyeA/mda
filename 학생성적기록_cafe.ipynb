```python
students = []
#학생 클래스 
class Student:
    #생성자 : 초기화작업
    def __init__(self, name, korean, english, math):
        self.name = name
        self.kor = korean
        self.eng = english
        self.mat = math
        self.total = korean + english + math
        self.avg = (korean + english + math) / 3
       
    #Setter : 외부에서 데이터를 필드에 입력
    def setName(self,name):
        self.name = name
    def setKor(self,kor):
        set.kor = kor
    def setEng(self,eng):
        self.eng = eng
    def setMat(self,math):
        self.mat = math
    
    #getter 외부에 데이터를 전달 
    def getName(self):
        return self.name
    def getKor(self):
        return self.kor
    def getEng(self):
        return self.eng
    def getMat(self):
        return self.mat
    def getTotal(self):
        self.total = self.kor + self.eng +self.mat
        return self.total
    def getAvg(self):
        self.avg = self.getTotal()/3
        return self.avg 
 # ========= class 끝 =====================
```

```python
# 외부 입력함수     
def input_s():
    name = input('이름을 입력하세요: ')
    korean = int(input('국어 점수를 입력하세요: '))
    english = int(input('영어 점수를 입력하세요: '))
    math = int(input('수학 점수를 입력하세요: '))
    
  
    student = Student(name, korean, english, math)
    students.append(student)
    
# 외부 전체출력함수 
def print_s():  
    for i in range(len(students)):
        print(students[i].getName(), end='\t')
        print(students[i].getKor(), end='\t')
        print(students[i].getEng(), end='\t')
        print(students[i].getMat(), end='\t')
        print(students[i].getTotal(), end='\t')
        print(students[i].avg)   

# 외부 검색함수         
def search_s():
    name2 = input('검색할 이름을 입력하세요: ')
    
    for i in range(len(students)):
        if(name2 == students[i].getName()):
            print(students[i].getName(), end='\t')
            print(students[i].getKor(), end='\t')
            print(students[i].getEng(), end='\t')
            print(students[i].getMat(), end='\t')
            print(students[i].getTotal(), end='\t')
            print(students[i].avg)

# 외부 수정함수 
def modify_s():
    name2 = input('수정할 이름을 입력하세요: ')
     
    for i in range(len(students)):
        if(name2 == students[i].getName()):
            print('수정할 과목을 선택하세요')
            print('1.국어 2.영어 3.수학')
              
            b = int(input())
               
            if(b == 1):
                students[i].setKor( int(input()))
            elif(b == 2):
                students[i].setEng( int(input()))
            elif(b == 3):
                students[i].setMat ( int(input()))
            else:
                print('잘못 입력하였습니다.')
                   
            students[i].getTotal()
            students[i].getAvg()

            # 외부 삭제함수 
def delete_s():       
        name2 = input('삭제할 이름을 입력하세요: ')
       
        for student in students:
            if(name2 == student.getName()):
                students.remove(student)
                
```

```python
# 외부 : 화면함수 
def main():
    print('선택하세요')
    print('1.학생정보입력 2.전체출력 3.검색 4.수정 5.삭제 6.종료')
   
    a = int(input())
   
    if(a == 1):
        input_s()
    elif(a == 2):
        print_s()
    elif(a == 3):
        search_s()
    elif(a == 4):
        modify_s()
    elif(a == 5):
        delete_s()
    elif(a == 6):
        return a
        
# start 부분 
while(True):
    
    ans = main()
    
    if(ans == 6):
        print('프로그램 종료')
        break
```
